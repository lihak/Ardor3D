# 강화학습 코드의 기본 구조

```python
# 환경 로직을 래핑하는 객체로 여기에서 gymnasium을 사용
env = load_env()

# 에이전트 정책(또는 value function)을 래핑하는 객체
# 파라미터 및 행동 생성 방법
agent = get_agent()

for episode in range(0, n_episodes):

    # 환경의 랜덤 시작
    obs, info = env.reset()

    done = False
    while not done:

        action = agent.get_action(obs)

        # 환경이 다음 상태로 전환되며 에이전트에게 보상을 제공
        next_obs, reward, terminated, truncated, info = env.step(action)

        # 에이전트의 파라미터를 조정
        agent.update(obs, action, reward, next_obs)

        done = terminated or truncated
        obs = next_obs
```
